```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("shuyangli94/food-com-recipes-and-user-interactions")

print("Path to dataset files:", path)
```

```python
import pandas as pd

recipes = pd.read_csv(f'{path}/PP_recipes.csv')
recipes
```

```python
train_df = pd.read_csv(f'{path}/interactions_train.csv')
train_df
```

```python
ingredient_map = pd.read_pickle(f'{path}/ingr_map.pkl')
ingredient_map
```

```python
# Use pandas.merge() to join the two dataframes
# This connects your training data to the recipe details
train_data_with_features = pd.merge(
    train_df,
    recipes,
    left_on='recipe_id',
    right_on='id'
)

train_data_with_features
```
